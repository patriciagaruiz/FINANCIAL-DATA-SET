```python
import random
import re

#Repository of possible words in the game
database_game = ['game','ball','doll','door','notebook','orange','house','beach','journey','pencil',
                 'vegetable','fruit','water','window','chocolate','banana','hurricane','flower',
                 'brother','rice','beans', 'potato', 'horse', 'fork','knife', 'paper', 'wood', 'sunglass']


def select_random_word(database_game):
    ''' This function choices only a word in the repository
    For example: random.choice(database_game) returns 'dool' from database_game'''

    return random.choice(database_game)

def find_hide_space(secret_word):
    ''' This function transforms each letter of secret_word (word aleatory chosen) in "underlines"
    The user see how many letters (undelines) the secret_word has
    For exemplo: secret_word is 'doll', hide_word_format =  _  _  _  _  '''
    
    hide_space = []
    for letter in secret_word:
        hide_space.append(' _ ')
    hide_word_format = '' 

    for i in hide_space:
        hide_word_format += ' _ '
    return hide_word_format

def input_letter(save_input_letter):
    ''' This function tests if the letter is valid and formats the letter to lower case saving the letter 
    in a repository for future checks
    If the letter was already input or not valid, returns a new input waiting for a valid letter
    For exemplo: letter = 9 returns new input wainting for a valid letter
    Return letter (valid_letter) and save_input_letter (save all letters that user input into a repository)''' 
    
    letter = input("Let's try! Input a letter: ")
    while letter.isdigit()==True or len(letter)>1:
        letter = input('Input 1 letter: ')
    letter = letter.lower()
    if letter in save_input_letter:
        print('Letter already input')
        letter, save_input_letter = input_letter(save_input_letter)
        letter = letter.lower()
        return letter, save_input_letter
    else:
 # Save into a list 'save_input_letter' all valid letters input by the user        
        save_input_letter.append(letter)
    return letter, save_input_letter        

def input_sequence_letter(save_input_letter, secret_word, secret_word_list, chance_lost, save_input_valid_letter):
    ''' This function has the logic of the game 
    Get a valid letter and consist if exists in the secret word
    Control the number of wrong choices. Game is over if the user input more than 3 wrong letters
    Save the valid letter that exists in secret word into a new repository (save_input_valid_letter)'''

    letter_valid, save_input = input_letter(save_input_letter)
    secret_word_list = secret_word_list
    new_display_secret_word = ''
# Verify if the valid letter exists in secret word and sum wrong choices
    if letter_valid not in secret_word:
        print("Letter doesn't exist")
        chance_lost += 1
        if chance_lost > 3:
            print('Sorry, game over :-( ')
            start_game()
        else:
            input_try = 0
            input_try = 4 - chance_lost
            print('You failed:', chance_lost, 'time(s)! You have more', input_try, 'try(es)')
    else:
# Save into a list 'save_input_valid_letter' only the valid letters that exist in secret word 
        if letter_valid not in save_input_valid_letter:
            save_input_valid_letter.append(letter_valid) 
            
    for letter in secret_word:
# Build a friendly display for secret word to be display
        if letter in save_input:
            new_display_secret_word += letter
        else:
            new_display_secret_word += ' _ '
            
# Verify if every letter into the secret word exist into a list that contain only the valid letters
    if set(save_input_valid_letter) == set(secret_word_list):
        print('You win!!! Congratulations! The secret word is:', new_display_secret_word)
        return start_game()
    else:
        print("The secret word is:",new_display_secret_word)
        input_sequence_letter(save_input_letter, secret_word, secret_word_list, chance_lost, save_input_valid_letter)



def start_game():
    ''' This function starts the game asking if the user want to play the game
    If the user is ready to go, this function calls a function that has
    the logic of the game (input_sequence_letter)
    If the user losts or wins, only in this function it is possible keep or leave
    the game '''
    
    while game_on == True:
        question_game = input('Are you ready? (y/n) ')
        if question_game == 'y' or question_game == 'Y':
            print('Good Luck!!!')
            chance_lost = 0
            save_input_letter = []
            secret_word_list = []
            save_input_valid_letter = []
#secret_word stores the target word
            secret_word = select_random_word(database_game)
            secret_word_list = re.findall('\w',secret_word)
            hide_word_display = find_hide_space(secret_word)
            print('Secret_word:', secret_word)
            print("Let's start!")
            print("The secret word is:",hide_word_display)
            input_sequence_letter(save_input_letter, secret_word, secret_word_list, chance_lost, save_input_valid_letter)
        elif question_game == 'n' or question_game == 'N':
            print("Let's try later")
            return
        else:
            print("Input y/n is required")
            return
        game_on = False
    print('Bye')
    

#start game!
game_on = True
start_game()
```
